# X gate

```python
import qiskit

from qiskit import *

qc= QuantumCircuit(1)  #creating one qubit

qc.x(0) #adding X gate to the 1st qubit

qc.draw()
```

# H gate

```python
qc.h(0) #H gate applied to 0 number qubit
qc.draw()
```

# Bloch Sphere

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector


state = Statevector(qc) #creating state vector for the circuit
plot_bloch_multivector(state) #bloch sphere
```

# Histogram

```python
from qiskit.visualization import plot_histogram

counts1 = {'1': 1}  #setting 1's value to 1

legend = ['First execution']

plot_histogram([counts1], legend=legend, color=['lightblue'], title="New Histogram")
```
